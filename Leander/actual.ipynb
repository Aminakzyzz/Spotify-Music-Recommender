```python
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd



data = pd.read_csv("data.csv")
```

```python
# X = data.select_dtypes(np.number)

# I removed 4 categorical attributes: ['id', 'artists', 'name', 'release_date']
allAttributes = [
    "valence",
    "year",
    "acousticness",
    "artists",
    "danceability",
    "duration_ms",
    "energy",
    "explicit",
    "id",
    "instrumentalness",
    "key",
    "liveness",
    "loudness",
    "mode",
    "name",
    "popularity",
    "release_date",
    "speechiness",
    "tempo",
]

attributesToKeep = [
    "valence",
    "year",
    "acousticness",
    "danceability",
    "duration_ms",
    "energy",
    "explicit",
    "instrumentalness",
    "key",
    "liveness",
    "loudness",
    "mode",
    "popularity",
    "speechiness",
    "tempo",
]
attributesToDrop = ["id", "artists", "name", "release_date"]
X = data.drop(attributesToDrop, axis=1, inplace=False)


K_NEIGHBORS = 10

kn = NearestNeighbors(n_neighbors=K_NEIGHBORS)
kn.fit(X)


inputData = dict(data.iloc[5])
inputData.pop("id")
inputData.pop("artists")
inputData.pop("name")
inputData.pop("release_date")
inputDataPd = pd.DataFrame([inputData])

# inp = X.iloc[5]
neighbours = kn.kneighbors(inputDataPd, K_NEIGHBORS, return_distance=False)
neighbours = neighbours[0]  # [neighbours]

print(neighbours)

print("Input:")
print(inputData)
print(f"{K_NEIGHBORS} Neighbours:")

songNames = []
for neigbour in neighbours:
    dic = dict(data.iloc[neigbour])
    print(neigbour, dic)
    outStr = f'{dic["name"]} by {dic["artists"]} from {dic["year"]}'
    songNames.append(outStr)
    
print(songNames)
```
